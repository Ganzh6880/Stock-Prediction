```python
import torch
import torch.nn as nn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# This relates to plotting datetime values with matplotlib:
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df = pd.read_csv('FB.csv',index_col=0,parse_dates=True)
len(df)
```

```python
df = df.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'],axis=1)
```

```python
plt.figure(figsize=(12,4))
plt.title('Facebook Price')
plt.ylabel('Price')
plt.grid(True)
plt.autoscale(axis='x',tight=True)
plt.plot(df['Close'])
plt.show()
```

```python
# Extract values from the source .csv file
y = df['Close'].values.astype(float)

# Define a test size
test_size = 180

# Create train and test sets
train_set = y[:-test_size]
test_set = y[-test_size:]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Instantiate a scaler with a feature range from -1 to 1
scaler = MinMaxScaler(feature_range=(-1, 1))
```

```python
train_norm = scaler.fit_transform(train_set.reshape(-1, 1))
```

```python
# Convert train_norm from an array to a tensor
train_norm = torch.FloatTensor(train_norm).view(-1)

# Define a window size
window_size = 60

# Define function to create seq/label tuples
def input_data(seq,ws):  # ws is the window size
    out = []
    L = len(seq)
    for i in range(L-ws):
        window = seq[i:i+ws]
        label = seq[i+ws:i+ws+1]
        out.append((window,label))
    return out

# Apply the input_data function to train_norm
train_data = input_data(train_norm,window_size)
len(train_data)  # this should equal 325-12-12
```

```python
# Display the first seq/label tuple in the train data
train_data[0]
```

```python
class LSTMnetwork(nn.Module):
    def __init__(self,input_size=1,hidden_size=50,num_layers=2,output_size=1):
        super().__init__()
        self.hidden_size = hidden_size
        
        # Add an LSTM layer:
        self.lstm = nn.LSTM(input_size,hidden_size,num_layers)
        
        # Add a fully-connected layer:
        self.linear = nn.Linear(hidden_size,output_size)
        
        # Initialize h0 and c0:
        self.hidden = (torch.zeros(1,1,self.hidden_size),
                       torch.zeros(1,1,self.hidden_size))

    def forward(self,seq):
        lstm_out, self.hidden = self.lstm(
            seq.view(len(seq),1,-1), self.hidden)
        pred = self.linear(lstm_out.view(len(seq),-1))
        return pred[-1]  # we only want the last value
```

```python
model = LSTMnetwork()

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

model
```

```python
def count_parameters(model):
    params = [p.numel() for p in model.parameters() if p.requires_grad]
    for item in params:
        print(f'{item:>6}')
    print(f'______\n{sum(params):>6}')
    
count_parameters(model)
```

```python
epochs = 50

import time
start_time = time.time()

model = model.cuda()

for epoch in range(epochs):
    
    i=1
    
    # extract the sequence & label from the training data
    for seq, y_train in train_data:
        
        seq = seq.cuda()
        y_train = y_train.cuda()
        
        # reset the parameters and hidden states
        optimizer.zero_grad()
        model.hidden = (torch.zeros(2,1,model.hidden_size).cuda(),
                        torch.zeros(2,1,model.hidden_size).cuda())
        
        y_pred = model(seq)
        
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        
        
        if i%100 == 0:
            print(f'Epoch {epoch+1:2} Batch {i} Loss:{loss.item():10.8f}')
        i+=1
        
    # print training result
    print(f'Epoch: {epoch+1:2} Loss: {loss.item():10.8f}')
    
print(f'\nDuration: {time.time() - start_time:.0f} seconds')
```

```python
future = 180

# Add the last window of training values to the list of predictions
preds = []
pred_data = train_norm.tolist()

model = model.cpu()

# Set the model to evaluation mode
model.eval()

for i in range(len(df)-window_size):
    seq = torch.FloatTensor(pred_data[i:i+window_size])
    with torch.no_grad():
        model.hidden = (torch.zeros(2,1,model.hidden_size),
                        torch.zeros(2,1,model.hidden_size))
        if i > len(df)-window_size-future-1:
            pred_data.append(model(seq).item())
        preds.append(model(seq).item())
```

```python
epochs = 100

model = model.cuda()

# set model to back to training mode
model.train()

# feature scale the entire dataset
y_norm = scaler.fit_transform(y.reshape(-1, 1))
y_norm = torch.FloatTensor(y_norm).view(-1)
all_data = input_data(y_norm,window_size)

import time
start_time = time.time()

for epoch in range(epochs):
    
    i=1
    
    # train on the full set of sequences
    for seq, y_train in all_data:  
        
        seq = seq.cuda()
        y_train = y_train.cuda()
        
        # reset the parameters and hidden states
        optimizer.zero_grad()
        model.hidden = (torch.zeros(2,1,model.hidden_size).cuda(),
                        torch.zeros(2,1,model.hidden_size).cuda())
        
        y_pred = model(seq)
        
        loss = criterion(y_pred, y_train)
        loss.backward()
        optimizer.step()
        
        if i%100 == 0:
            print(f'Epoch {epoch+1:2} Batch {i} Loss:{loss.item():10.8f}')
        i+=1
        
    # print training result
    print(f'Epoch: {epoch+1:2} Loss: {loss.item():10.8f}')
    
print(f'\nDuration: {time.time() - start_time:.0f} seconds')
```

```python
window_size = 60
future = 60
L = len(y)

preds = []
pred_data = y_norm.tolist()

model = model.cpu()

model.eval()
for i in range(len(df)):  
    seq = torch.FloatTensor(pred_data[i:i+window_size])
    with torch.no_grad():
        # Reset the hidden parameters here!
        model.hidden = (torch.zeros(2,1,model.hidden_size),
                        torch.zeros(2,1,model.hidden_size)) 
        if i > len(df)-future-1:
            pred_data.append(model(seq).item())
        preds.append(model(seq).item())

# Inverse-normalize the prediction set
true_predictions = scaler.inverse_transform(np.array(preds).reshape(-1, 1))
print(true_predictions)

# PLOT THE RESULT
# Set a data range for the predicted data.
# Remember that the stop date has to be later than the last predicted value.
print(type(x))
df_new = np.arange('2020-04-24', '2020-06-23', dtype='datetime64[D]').astype('datetime64[D]')
x = np.concatenate((x,df_new), axis=None)
print(x)

plt.figure(figsize=(12,4))
plt.title('Beer, Wine, and Alcohol Sales')
plt.ylabel('Sales (millions of dollars)')
plt.grid(True)
plt.autoscale(axis='x',tight=True)
plt.plot(df['Close'])
plt.plot(x,true_predictions)
plt.show()
```
